Michael Muschitiello // Shannon Entropy of S&P500


This notebook loads and cleans multiple CSV files containing historical S&P 500 constituent data.  
It then computes Shannon Entropy for each year’s index weighting, normalizes the entropy,  
and plots the evolution of normalized entropy across the available time range.

Summary of Steps:
1. Load data from CSV files for different years.  
2. Perform basic cleaning (removing unnecessary characters and handling missing data).  
3. Compute each year’s entropy based on weights.  
4. Normalize entropy by dividing by the maximum possible entropy (log of number of constituents).  
5. Plot the normalized entropy to visualize how market concentration evolves over time.

Computing Shannon, Max, and Normalized Entropy based on weights of each years constituent list

```python
def compute_shannon_entropy_for_yearly_weights(yearly_weights, start_year=2005):
    """
    For each DataFrame in yearly_weights (each having columns 'BLOOMBERG TICKER' and 'INDEX WEIGHT'),
    this function computes Shannon Entropy, Max Entropy, and Normalized Entropy.
    Returns a list of dictionaries with the computed values for each year.
    
    Args:
        yearly_weights (list of pd.DataFrame): A list of DataFrames with 'BLOOMBERG TICKER' and 'INDEX WEIGHT'
        start_year (int): The year associated with the first DataFrame. Defaults to 2015.
    
    Returns:
        list of dict: Each dict has keys {"year", "shannon", "max", "normalized"}.
    """
    results = []
    for idx, df in enumerate(yearly_weights, start=1):
        year = start_year + idx - 1
        
        temp_df = df.copy()

        # rename columns 
        temp_df.rename(columns={
            'BLOOMBERG TICKER': 'Ticker',
            'INDEX WEIGHT': 'Weight'
        }, inplace=True)
        
        # set ticker as the index
        temp_df.set_index('Ticker', inplace=True)
        
        total_weight = temp_df['Weight'].sum()
        if total_weight == 0:
            print(f"Year {year}: All weights are zero, skipping entropy calculation.")
            continue
        
        temp_df['Fraction'] = temp_df['Weight'] / total_weight
        
        # shannon entropy
        p = temp_df['Fraction'].values
        shannon_entropy = -np.sum(p * np.log(p))
        
        # max entropy (log of number of tickers)
        max_entropy = np.log(len(p))
        
        # Compute normalized entropy
        normalized_entropy = shannon_entropy / max_entropy if max_entropy > 0 else np.nan
        
        results.append({
            "year": year,
            "shannon": shannon_entropy,
            "max": max_entropy,
            "normalized": normalized_entropy
        })
        
    return results


entropy_results = compute_shannon_entropy_for_yearly_weights(yearly_weights, start_year=2005)
df_entropy = pd.DataFrame(entropy_results)

# plot
plt.figure(figsize=(18, 6))
plt.plot(df_entropy['year'], df_entropy['normalized'], marker='s', label='Normalized Entropy')
plt.title('Normalized Entropy Progression by Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Entropy')
plt.grid(True)
plt.legend()
plt.show()

yearly_weights = [const_05, const_06, const_07, const_08, 
                  const_09, const_10, const_11, const_12, 
                  const_13, const_14, const_15, const_16, 
                  const_17, const_18, const_19, const_20, 
                  const_21, const_22, const_23, const_24, const_25]

compute_shannon_entropy_for_yearly_weights(yearly_weights)
```
